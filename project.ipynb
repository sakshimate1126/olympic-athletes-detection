```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.countplot(x=df['Age'],order=df['Age'].value_counts().index.sort_values())
plt.xlabel('Age',fontsize=30)
plt.ylabel('Total Medals',fontsize=30)
plt.title('Total Medals vs Age',fontsize=40);
```

```python
plt.figure(figsize=(10,12))
plt.pie(df['Sport'].value_counts(),labels=df['Sport'].value_counts().index,autopct="%1.2f%%")
plt.title("sports")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(x=df['Country'],order=df['Country'].value_counts().index.sort_values(),palette='dark')
plt.xticks(rotation='vertical')
plt.xlabel('Country',fontsize=30)
plt.ylabel('Gold Medals',fontsize=30)
plt.title('Gold Medals vs Country',fontsize=40);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(y=df['Athlete'],order=df['Athlete'].value_counts().index[0:10],palette='deep')

plt.xlabel('Total Medals',fontsize=12)
plt.ylabel('Athlete',fontsize=12)
plt.title('Athlete vs Total Medals ',fontsize=16);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(y=df['Country'],order=df['Country'].value_counts().index[0:10],palette='magma')

plt.xlabel('Country',fontsize=30)
plt.ylabel('Age',fontsize=30)
plt.title('Age vs Country',fontsize=40);
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df[df['Athlete']!=0]['Athlete'].value_counts().index[0:5],
           y=df['Athlete'].value_counts()[0:5],palette='CMRmap')
plt.xticks(rotation='vertical')
plt.xlabel('Athlete',fontsize=12)
plt.ylabel('Silver Medals',fontsize=12)
plt.title('Silver Medals vs Athlete',fontsize=14);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.lineplot(x=df['Age'],order=df['Age'].value_counts().index.sort_values())
plt.xlabel('Age',fontsize=30)
plt.ylabel('Total Medals',fontsize=30)
plt.title('Total Medals vs Age',fontsize=40);
```
